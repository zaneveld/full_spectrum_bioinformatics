# Setting up the Windows Subsystem for Linux and Miniconda 

<img src="./resources/1_conda_conceptual_overview.png" width="600"  description="A series of rectangles showing the relationship of the Windows Subsystem for Linux, BASH, and conda virtual environments. An outer box labeled Powershell contains WSL which contains a Linux command line interface, which contains two different conda environments, which in turn contain different software packages.">

Most bioinformatic software is not developed to run on Windows machines. This is in part because larger analyses often need to move onto a computing cluster, and currently most academic computing clusters use Linux. 

Thankfully, there are now some relatively seamless ways to run Linux software on Windows using the Windows Subsystem for Linux (WSL). This will allow you to switch to using Linux within Windows PowerShell, and from there allow you to use command line commands to download and install scientific software for sequence alignment, phylogenetics, microbiome analysis or other common tasks.

The main thing to be aware of is that because the Windows Subsystem for Linux is distinct from your usual Windows operating system, programs like conda that you installed on Windows will generally not be available within WSL. Instead, we'll have to install them from the command line.

In this chapter, we'll cover the basics of how to set up WSL on Windows. Following that, we will use command line commands to download and install Miniconda. This is a miniature version of the `conda` software that comes with Anaconda python. Many, though sadly not all, scientific software packages can be installed using simple commands given to Miniconda.

## Installing the Windows Subsystem for Linux

## Installing WSL

First, open Windows PowerShell. Then, type the following:

`wsl --install`

For our purposes the default options are just fine, but if you ever need specific versions of Linux for a project or want to learn more background about WSL, additional guidance is available from Microsoft [here](https://learn.microsoft.com/en-us/windows/wsl/install).  

You may need to reboot as part of the installation process. 

You may also have WSL already installed. If so you will get an error that says something like :
```
Error code: Wsl/InstallDistro/ERROR_ALREADY_EXISTS
PS C:\Windows\System32>
```

In that case you can simply activate WSL by typing `WSL`

### Make sure WSL is active

In a clean PowerShell window, type `wsl` to make sure WSL is active. Then do the next steps of the installation in that window with wsl active.

## Install Miniconda into WSL

Miniconda is effectively a smaller version of Anaconda Python that uses less disk space. The official guidance for how to install Miniconda into WSL is available [here](https://docs.conda.io/projects/conda/en/latest/user-guide/install/index.html). **However**, in practice there are some tricks to getting this to work in WSL. Specifically, if you just download and install Miniconda through the graphical user interface, you'll be in Windows rather than WSL, so you will have Miniconda in Windows but not in WSL.

Instead, the cleanest way to install Miniconda on Windows is to do everything from the command line. Thankfully, Linux comes with convenient programs for downloading and running installers for software. As long as we run everything from the command line, *with WSL activated*, then Miniconda can be set up in WSL with just a couple of commands.

### Download the Miniconda Installer using `wget`

**Before we download Miniconda, be sure you are in WSL.** If in doubt, restart your computer, open a new PowerShell terminal, and type `wsl`. You should see some text indicating that this terminal session is now in wsl. 

`wget` is a Linux program for downloading files. It is commonly used when setting up new software or downloading datasets from online sources. In this case we want to point `wget` at the latest installer for 
Miniconda. This installer is a shell script with the file extension `.sh`. To download it, we  pass the address where the latest version of Miniconda is hosted on the Anaconda website, and pass that as a positional argument to wget, as shown below:

`wget https://repo.anaconda.com/miniconda/Miniconda3-latest-Linux-x86_64.sh`

This will download the install `.sh` file. 

**NOTE: if you ever wanted to download a different file using `wget`, you would just type `wget` then a space, then the full web address of the file you wanted to download.**

### Install Miniconda 



Now that we have the `.sh` file that will install Minconda, we need to run it to set up Miniconda. We can either do this interactively or in batch mode (see below). The `bash` command is used to run scripts of other command line commands written out in `.sh` files. We want to run all the commands in the Miniconda installer `.sh` file, so we will call `bash` and pass the name of that file as a positional argument.

#### Installing Miniconda Interactively

`bash Miniconda3-latest-Linux-x86_64.sh`

If you ran the installer in the default way, will see several interactive prompts from the installer that you will need to respond to for installation to finish. The first will ask about whether you accept the Miniconda license. Follow the instructions onscreen to complete the installation. You may have to hit Enter repeatedly to read through the license.

For educational and academic users Miniconda is currently free (this is not legal advice; see https://www.anaconda.com/legal/terms/terms-of-service for the official current terms). If you're at a company with >200 employees, or doing research that is funded by such a company you may wish to check out free alternatives like [Miniforge](https://github.com/conda-forge/miniforge). 

To agree to the liscence, type `yes` and hit Enter.

You will also be asked about the default installation location (the default is good unless you have a specific reason to do something different).

Finally, you will be prompted about initializing conda. You want to say `yes` to this (this is important to make sure Miniconda will be available).

#### Installing Miniconda in batch mode

If you prefer to automatically accept all prompts, rather than interactively accepting them, you can run the installer with the `-b` flag parameter:

`bash Miniconda3-latest-Linux-x86_64.sh -b`

 The `-b` flag parameter automatically accepts all prompts (if you work at a corporation, be sure to understand the Anaconda licence before doing this for a work project).
 
 
### Restart your Command Line interface

After installation is done, you'll need to close your PowerShell window, open a new one, restart WSL by typing `wsl`. If conda was successfully installed iwth default options you should see the word `(base)` appear before your cursor. Conda allows you to set up what are known as virtual environments. These are basically bubbles of compatible software - with what you install in one virtual environment not interfering with incompatible software in other virtual environments. The `(base)` means you are currently in the base or main conda environment. 

You can also check that Miniconda is installed properly by typing:

`conda --help`

## Using conda

There are two main uses for Miniconda now that you have it installed: installing other software and setting up virtual environments to prevent software packages from interfering with each other.



### Installing other software

Many scientific software packages can be installed using conda. The software and it's dependencies (other software or libraries it needs to run) are found by searching *channels*. These can include the default channel as well as others such as `conda-forge`, `bioconda` etc. 

For software or python libraries that are in the default channel, the install command is often some version of:

`conda install <name of your package goes here without the '<' or '>'>` 

For other software, you may need to first configure conda channels. For example, you can configure some common conda channels with these commands:

`conda config --add channels defaults`
`conda config --add channels bioconda`
`conda config --add channels conda-forge`

For example, the `muscle` multiple sequence alignment software currently say it can be installed using the command:

`conda install muscle`

In some cases, different versions of software are available in different channels. In the cases that this matters you can specify which version you get using a colon after the name of what you want to install:

`conda install muscle:bioconda`

**NOTE:** Whatever virtual environment you are in (see below) will be where software is installed. So if you want it to go into a specific virtual environment, you want to activate that virtual environment before installing software. 

### Setting up virtual environments

Conda can create [virtual environments](https://docs.conda.io/projects/conda/en/stable/user-guide/tasks/manage-environments.html) for your projects. This basically means that you can set up 'bubbles' of software that are compatible with each other. Why does that matter? I found out the hard way before conda was available. Let's say you want to install one piece of scientific software that requires a new version of python, but already have older software that can only work on older versions of python. This creates a **conflict** between the requirements for one piece of software and the other. 

Virtual environments provide a solution: you can install the newest version of python and one piece of software into one virtual environment and an older version of python into another.

Generally, it's a good practice to create a virtual environment for each project (or set of projects that use similar techniques). Some complex software packages like qiime2 *require* that you create a virtual environment to run them, because they in turn run many other software packages that have specific requirements.

When you set up a conda environment, you can specify what you want installed into it. Or you can create a simple empty environment and install each piece of software into it.

#### Create a new conda environment

Within conda you can create a virtual environment to set up software within. This is done with the command `conda create` followed by information about the name of the environment and what software you want installed into it. Here are a couple of [examples](https://docs.conda.io/projects/conda/en/stable/user-guide/tasks/manage-environments.html) modified from the conda documentation (you don't need to run them right now):

`conda create -n python3_env python=3.9`

This would create a python environment and install python 3.9 into it. 

In some cases, complex environments are set up by specifying a `.yml` file with all the software and libraries you want to set up listed. Here is a more complex example from the QIIME2 software package for microbiome analysis:

`conda env create -n qiime2-amplicon-2024.10 --file https://data.qiime2.org/distro/amplicon/qiime2-amplicon-2024.10-py310-linux-conda.yml`
`

#### List existing virtual environments

If like me you always forget what the name(s) of your conda environments are, you can list them with:

`conda env list`

#### Activate a conda environment

Once you've reminded your
Then just pick the one you want to use from the list and activate it with `conda activate <name of environment>'.

So for example the QIIME2 conda environment above would be activated (if installed) with:

`conda activate qiime2-amplicon-2024.10`

Note that this command only works within WSL. If you rebooted, you'd have to open a new PowerShell window, type `wsl` to get into Linux, then type the command to fire up your virtual environment.
